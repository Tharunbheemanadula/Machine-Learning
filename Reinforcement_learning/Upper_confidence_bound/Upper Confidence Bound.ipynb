# Upper Confidence Bound

## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing dataset

```python
data_set=pd.read_csv('Upper Confidence Bound Facebook.csv')
```

## Implementation

```python
import math as m
t=15000

num_ads=10
ad_selected=[]
num_of_selection=[0]*num_ads
sum_of_rewards=[0]*num_ads
total_rewards=0

for n in range(0,t):
    avg_reward=0
    max_upper_confidence_bound=0
    for i in range(0,num_ads):
        if(num_of_selection[i]>0):
            avg_reward=sum_of_rewards[i]/num_of_selection[i]
            delta_i=m.sqrt(3/2*(m.log(n+1)/(num_of_selection[i])))
            upper_confidence_bound=avg_reward+delta_i
        else:
            upper_confidence_bound=1e300
        if(upper_confidence_bound>max_upper_confidence_bound):
            max_upper_confidence_bound=upper_confidence_bound
            ad=i
        ad_selected.append(ad)
        num_of_selection[ad]=num_of_selection[ad]+1
        temp_reward=data_set.values[n,ad]
        sum_of_rewards[ad]=sum_of_rewards[ad]+temp_reward
        total_rewards+=temp_reward
```

## Histogram Visualization

```python
plt.hist(ad_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
