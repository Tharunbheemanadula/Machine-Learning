# Thomson Sampling

## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing dataset

```python
dataset = pd.read_csv('Thomson Sampling Facebook Ads.csv')
```

## Implementation

```python
import math
import random
T=15000
num_ads=10
ads_selected=[]
num_rewards1=[0]*num_ads
num_rewards0=[0]*num_ads
total_rewards=0
for n in range(0,T):
    ads=0
    max_random=0
    for i in range(0,num_ads):
        random_beta=random.betavariate(num_rewards1[i]+1,num_rewards0[i]+1)
        if(random_beta>max_random):
            max_random=random_beta
            ad=i
        ads_selected.append(ad)
        reward=dataset.values[n,ad]
        if reward==1:
            num_rewards1[ad]+=1
        else:
            num_rewards0[ad]+=1
        total_rewards+=reward
        
```

## Histogram Visualization

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
